```python
import pandas as pd
```

```python
import re
```

```python
df=pd.read_csv("Restaurant_Reviews.tsv","\t")
```

```python
df.head()
```

```python
#!pip install nltk
import nltk
#nltk.download('stopwords')
nltk.corpus.stopwords.words('english')
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
def preprocess(msg):
    msg=msg.lower()
    msg=re.sub('[^a-z]',' ',msg)
    final=[ps.stem(p) for p in msg.split() if p not in nltk.corpus.stopwords.words('english')]
    return " ".join(final)
```

```python
df['Review']=df['Review'].apply(preprocess)
```

```python
preprocess('Crust is not good.')
```

```python
from sklearn.feature_extraction.text import CountVectorizer

bow=CountVectorizer()
```

```python
bowtable=bow.fit_transform(df['Review'])
```

```python
X=bowtable.toarray()
```

```python
X
```

```python
X.shape
```

```python
y=df['Liked']
```

```python
y.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
params = {'max_depth':[2,4,10],'min_samples_leaf':[2,4,8],'random_state':[42,1,10,20,35],'max_leaf_nodes':[50,100]}
```

```python
grid_cv = GridSearchCV(
estimator=RandomForestClassifier(),param_grid=params,cv=5,verbose=2,n_jobs=3)
grid_cv.fit(X,y)
```

```python
grid_cv.best_params_
```

```python
model=RandomForestClassifier(max_depth=10,min_samples_leaf=2,random_state=10,max_leaf_nodes=50)
model.fit(X,y)
model.score(X,y)
```
